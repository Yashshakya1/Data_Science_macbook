# Kernel Density Estimation

* it is a way to estimate the probability distribution of continuous data.
* KDE shows how data is spread smoothly, instead of using bars like a histogram.

* Draws a smooth curve
* Shows where values are concentrated

# ðŸ”µ How KDE works (very simple)

* Take each data point

* Put a small smooth curve (kernel) on it

* Add all curves together

* Get one smooth density curve

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
ds = sns.load_dataset('penguins')
top = ds.head()
species = ds['species'].value_counts()
island = ds['island'].value_counts()
sex = ds['sex'].value_counts()
print(top)
print(f'{species}\n {island}\n{sex}')
```

```python
sns.kdeplot(data = ds,x='body_mass_g',hue = 'species',fill = True) # fill is used to fill the color inside the line
```
