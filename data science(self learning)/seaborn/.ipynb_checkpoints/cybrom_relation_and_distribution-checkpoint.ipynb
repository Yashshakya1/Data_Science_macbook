```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ds = sns.load_dataset('flights')
ds
```

```python
sns.scatterplot(data = ds,x='year',y = 'passengers')
```

```python
sns.relplot(data = ds,x='year',y = 'passengers',kind = 'scatter')
```

```python
sns.relplot(data = ds,x='year',y = 'passengers',kind = 'line')
```

```python
sns.lineplot(data = ds,x='year',y = 'passengers')
```
