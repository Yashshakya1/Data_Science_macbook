```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
ds = sns.load_dataset('penguins')
top = ds.head()
species = ds['species'].value_counts()
island = ds['island'].value_counts()
sex = ds['sex'].value_counts()
print(top)
print(f'{species}\n {island}\n{sex}')
```

# Rugplot 

```python
sns.rugplot(data = ds,x = 'body_mass_g')
```

```python
sns.rugplot(data = ds,x = 'body_mass_g',hue = 'species',palette='Set2')
```

```python
sns.rugplot(data = ds,y = 'body_mass_g',hue = 'species',palette='Set2',height=0.3)
```

#  What is pairplot in Seaborn?

* A pairplot is a figure-level plot in Seaborn used to visualize relationships between multiple numerical variables at once.
   - It creats__
  
* Scatter plots for every pair of variables
* Histogram or KDE on the diagonal (distribution of each variable)

| `diag_kind` | Meaning             |
| ----------- | ------------------- |
| `"hist"`    | Histogram (default) |
| `"kde"`     | KDE distribution    |


```python
sns.pairplot(data = ds)
```

```python
sns.pairplot(data = ds,diag_kind='kde')
```

* i in hue this pairplot automatically kde in the digonal part 
* with using any kind_diagonal parameter 

```python
sns.pairplot(ds,hue = 'island', palette='Set1')
```
