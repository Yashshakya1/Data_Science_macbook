```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
ds = sns.load_dataset('penguins')
top = ds.head()
species = ds['species'].value_counts()
island = ds['island'].value_counts()
sex = ds['sex'].value_counts()
print(top)
print(f'{species}\n {island}\n{sex}')
```

```python
ds.describe()
```

```python
columns = ['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']
ds[columns].corr()
```

# What is Heatmap
* A heatmap in Seaborn is a visualization that represents data values in a matrix form using color to show intensity or magnitude.

| Parameter    | Meaning            |
| ------------ | ------------------ |
| `data`       | Matrix / DataFrame |
| `annot=True` | Show values        |
| `cmap`       | Color style        |
| `cbar`       | Show color bar     |
| `linewidths` | Line between cells |


# why use Heatmap
1. visulize the confusion matrices
2. to find the correlation between the features
3. highlight missing values 

```python
sns.heatmap(data=ds[columns].corr())
```

```python
sns.heatmap(data = ds[columns].corr(),annot = True) # annot = True showing the correlation values
```

```python
sns.heatmap(data = ds[columns].corr(),annot = True ,cmap = 'Blues')
```

```python
sns.heatmap(data = ds[columns].corr(),annot = True ,cmap = 'Blues',vmin = -0.2) # vmin = -0.2 range recreate
```

```python
sns.heatmap(data = ds[columns].corr(),annot = True ,cmap = 'Blues',linewidths=3,linecolor = 'black')
# linewidths create the space between the box 
# linecolour fill the colour in space porstion
```

# 2 Missing values visualization

```python
sns.heatmap(ds.isnull(), cbar= False)
```

# 3 Confusion matrix (ML)

```python
import numpy as np

cm = np.array([
    [50, 5],
    [3, 42]
])

sns.heatmap(cm, annot=True, fmt='d')
```

50  True Negatives

5  False Positives

3  False Negatives

42  True Positives
