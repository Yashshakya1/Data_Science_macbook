```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
ds = sns.load_dataset('penguins')
top = ds.head()
species = ds['species'].value_counts()
island = ds['island'].value_counts()
sex = ds['sex'].value_counts()
print(top)
print(f'{species}\n {island}\n{sex}')
```

```python
sns.boxplot(data=ds,x= 'species',y='body_mass_g')
```

```python
sns.boxplot(data=ds,x= 'species',y='body_mass_g',hue='sex')
```

# violinplot

* A violin plot shows how the data is distributed and where values are concentrated.

#  Interview one-line answer
* A violin plot is a combination of a box plot and a density plot that shows the full distribution of data across categories.

```python
sns.violinplot(data=ds,x= 'species',y='body_mass_g',hue='sex')
```

```python
sns.violinplot(data=ds,x= 'species',y='body_mass_g',hue='sex',split=True)
```

```python
sns.swarmplot(data = ds, x = 'species',y= 'body_mass_g',size = 3,hue = 'island',)
```
