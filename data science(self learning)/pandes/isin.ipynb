isin is a pandas fuction is used for fillter rows,check membership ,or match the multiple values in series or dataframe coloumns

```python
import seaborn as sns 
ds = sns.load_dataset('iris')
ds
```

```python
ds['petal_width'].isin([0,2,0,3,43,324,342,54])
```

```python
ds[ds['petal_width'].isin([0.2,0.3])]
```

```python
ds[ds['petal_width'].isin([0.2,0.3,0.4])] = 100
```
