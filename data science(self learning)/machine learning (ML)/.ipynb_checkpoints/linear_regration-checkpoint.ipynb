# ðŸ”¥ Difference Between array() and asarray()
* âœ” np.array()
* Always creates a new copy of the data (unless told otherwise).
* âœ” np.asarray()
* Creates a view, not a copy, whenever possible â†’ faster.

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
# data
x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([10,30,50,70,90])

# model
model = LinearRegression()
model.fit(x,y)
# Prediction
y_pred = model.predict(x)
# plotting
plt.scatter(x,y, color = 'red', label = 'Actual data')
plt.plot(x,y_pred ,color = 'green' ,label ='fit line')
plt.title('example of linear regression')
plt.xlabel('Study hours')
plt.ylabel('Marks')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Training data
x_axis = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y_axis = np.array([10, 30, 50, 70, 90])

# Create model and fit
model = LinearRegression()
model.fit(x_axis, y_axis)

# Take user input
hours = float(input('Enter your study hours: '))

# Predict
pred = model.predict([[hours]])

# Output
print(f'Study hours: {hours} â†’ Predicted score: {pred[0]:.2f}')
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

ds = pd.read_csv('pokemon_stats_2025.csv')
ds
```

# Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ds['name'] = le.fit_transform(ds['name'])
ds['type_1'] = le.fit_transform(ds['type_1'])
ds['type_2'] = le.fit_transform(ds['type_2'])
# for col in ['name','type_1','type_2']:
#     ds[col] = le.fit_transform(ds[col])
```

```python
ds
```

# extract the input feature x and traget value y

```python
x = ds.iloc[:,:-1]
y = ds['speed']

print('this is input feature that was x :-',x)
print('the target value :- \n' , y)
```

```python
# the shape of this dataset is 

print("Shape of X:", x.shape)
print("Shape of y:", y.shape)
print("Dataset shape:", ds.shape)
```

# check correlation

```python
corr_matrix = ds.select_dtypes(include = [int,float]).corr()
sns.heatmap(corr_matrix,annot = True)
```

# Feature salection

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x = pd.DataFrame(ss.fit_transform(x) , columns = x.columns)
```

```python
x
```

# split the data on the basis off test and train 

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.2, random_state=2)
print('this shape is before the splitting the data :-',x.shape)
print('After splitting the data :- ',x_train.shape)
print(x_train)
```

# creatign the LinearRegression Model 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
# -------------------------------------------------------
# 9. PREDICTION SYSTEM
# -------------------------------------------------------

def predict_speed(name, type_1, type_2, hp, attack, defense, sp_attack, sp_defense, height_m, weight_kg, total):

    row = [[name, type_1, type_2, hp, attack, defense, sp_attack, sp_defense, height_m, weight_kg, total]]

    input_df = pd.DataFrame(row, columns=x.columns)

    input_scaled = ss.transform(input_df)

    prediction = lr.predict(input_scaled)

    return prediction[0]




print("\n------------------- PREDICTION -------------------")


pred = predict_speed(
    name=10,
    type_1=3,
    type_2=5,
    hp=80,
    attack=95,
    defense=85,
    sp_attack=100,
    sp_defense=90,
    height_m=1.7,
    weight_kg=72,
    total=540
)

print(pred)
```

# find the m and c value 

```python
print('this is the value of C is :-',lr.intercept_)
```

```python
print('This is the value of M')
for col,coef in zip(x.columns,lr.coef_):
    print(f'{col} :- {coef}')
```

# Model Accuracy

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
y_pred = lr.predict(x_train)
print(y_test.shape)
print(y_pred.shape)
print("RÂ² Score:", r2_score(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Predict on test data
y_pred = lr.predict(x_test)

# R2 Score
print("RÂ² Score:", r2_score(y_test, y_pred))

# MAE
print("MAE:", mean_absolute_error(y_test, y_pred))

# MSE
print("MSE:", mean_squared_error(y_test, y_pred))

# RMSE
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```
