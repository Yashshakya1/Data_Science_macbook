# linspace 
linspace is a numpy fuction which is used for creating an array with evenly space number between a start and end vlaues 

this fuction is similar to the arange fuction but insted of step size

```python
ar = np.linspace(1,10)
print(ar)
```

```python
print(np.linspace(1,10,5 , endpoint = False))
```

```python
print(np.linspace(0,1,100))
```

```python
print(np.linspace(0,10,5, retstep = True))
```

```python
print(np.linspace(0,10,5,retstep = True,endpoint = False))
```
