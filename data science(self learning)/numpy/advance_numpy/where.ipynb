# Where function 

✅ What is np.where() in NumPy?

np.where() is a conditional function that allows you to:

* ✔️ Find positions (indexes) where a condition is True
* ✔️ Replace values based on a condition
* ✔️ Apply if–else logic to arrays

```python
import numpy as np
arr = np.arange(1,11)
print(arr)
```

# find position(index) where the condtion is true 

```python
np.where(arr > 5)
```

```python
ar = np.random.randint(0,10,10)
print(ar)
result = np.where(ar>5)
print(result)
```

```python
print(ar[np.where(ar>5)])
```

```python
# replace the value based of condtion 
print(np.where(arr > 5,1,0))
print(np.where(arr > 5,1,arr))
```

```python
np.where(arr == 5,100,arr)
```

```python
np.where((arr > 2) & (arr < 5),'yes','n')
```

```python
a = np.random.randint(1,10,2)
b = np.random.randint(1,10,2)
ab = np.concatenate((a,b), axis = 0)
print(ab)
```

```python
print(np.split(ab,2))
print(ab)
```

```python
sort = np.sort(ab)
print(sort)
decending = sort[::-1]
print(decending)
```
