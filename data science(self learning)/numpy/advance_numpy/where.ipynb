# Where function 

✅ What is np.where() in NumPy?

np.where() is a conditional function that allows you to:

* ✔️ Find positions (indexes) where a condition is True
* ✔️ Replace values based on a condition
* ✔️ Apply if–else logic to arrays

```python
import numpy as np
arr = np.arange(1,11)
print(arr)
```

# find position(index) where the condtion is true 

```python
np.where(arr > 5)
```

```python
# replace the value based of condtion 
print(np.where(arr > 5,1,0))
print(np.where(arr > 5,1,arr))
```

```python
np.where(arr == 5,100,arr)
```

```python
np.where((arr > 2) & (arr < 5),'yes','n')
```
