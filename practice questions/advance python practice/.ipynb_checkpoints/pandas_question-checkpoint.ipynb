Q.2  How to create a series from a list, numpy array and dict?


```python
li = ['a','b','c','d','e']
print(pd.Series(li))

ar = np.array(['dsfs','fd','dsf'])
print(pd.Series(ar))

dic = {'name':['yash','suraj','rani','tushar'],
      'marks':[99,78,98,00,89]}
pd.Series(dic)
```

Q.3 How to convert the index of a series into a column of a dataframe?

# reset_index()

```python
series = pd.Series([3,4,5], index = ['a','b','c'])
print(series)
series.reset_index()
```

Q.4 How to get the items not common to both series A and series B?

```python
s = pd.Series([1,2,3,4,5])
s1 = pd.Series([1,5,90,32,67])
pd.concat([s,s1])
```

```python
pd.concat([s[~s.isin(s1)],s1[~s1.isin(s)]])
```
