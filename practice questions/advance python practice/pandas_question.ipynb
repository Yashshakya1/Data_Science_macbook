Q.2  How to create a series from a list, numpy array and dict?

Q.4 How to get the items not common to both series A and series B?

```python
s = pd.Series([1,2,3,4,5])
s1 = pd.Series([1,5,90,32,67])
pd.concat([s,s1])
```

```python
pd.concat([s[~s.isin(s1)],s1[~s1.isin(s)]]) # this line is showing the not common element 
```

```python
pd.concat([s[s.isin(s1)],s1[s1.isin(s)]]) # show the common element 
```
