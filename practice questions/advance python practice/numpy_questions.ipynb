Q.1 Write a NumPy program to create a 3x3 matrix with values ranging from 2 to 10.

Q.2 Write a NumPy program to create a null vector of size 10 and update the sixth value to 5.

```python
zero = np.zeros(10, dtype = int)
zero[4] = 5
print(zero)
```

Q.3 Write a NumPy program to reverse an array (the first element becomes the last).

```python
ar = np.random.randint(1,10,6)
reverse = ar[::-1]
print(ar)
print(reverse)
```

Q.4 Write a NumPy program to create an 8x8 matrix and fill it with a checkerboard pattern.

```python
m = np.zeros((8,8),dtype = int)
print(m)
```

```python
matrix = np.zeros((8,8) , dtype = int)
matrix[::2,1::2] = 1
matrix[1::2,::2] = 1
print(matrix)
```

Q.6 Write a NumPy program to find common values between two arrays.
Expected Output:
Array1: [ 0 10 20 40 60] Array2: [10, 30, 40]

```python
array1 = np.array([43,34,34,54,43])
array2 = np.array([34,43])
np.intersect1d(array1,array2)
```

Q.8 Write a NumPy program to find the union of two arrays. Union will return a unique, sorted array of values in each of the two input arrays.
Array1: [ 0 10 20 40 60 80] Array2: [10, 30, 40, 50, 70]

```python
arr1 = np.array([0,54,20,90,60,80])
arr2 = np.array([20,30,40,50,10])
union = np.union1d(arr1,arr2)
print(union)
```

```python
reverse = union[::-1]
print(reverse)
```
